```python
%pylab --no-import-all inline
import numpy as np
import nengo
import symbols
import constructs
import nengo.networks as networks
import matplotlib
import matplotlib.pyplot as plt
import dot_product
#from collections import namedtuple
```

```python
D = 12
n_per_d = 24
N = D * n_per_d
```

```python
# Base Symbols
_ZERO, _F = symbols.initial_values(D)
model = nengo.Network()
numeral_table = dict(symbols.numerals(D, 20, zero=_ZERO, f=_F))
def hold_for_t(t, value):
    '''yields function that outputs value for time <= t'''
    default = np.zeros_like(value)
    def f(tau):
        return value if tau <= t else default
    return f

def symbolic_decode(numeral_table, x):
    similarities = {k: np.dot(v.v, x.T) for k, v in numeral_table.iteritems()}
    return np.vstack([d for _, d in sorted([(k, v) for k, v in similarities.iteritems()])])
```

## The Model

The model has four inputs:
1. "A" the numeral symbol

```python
with model:
    A_in = nengo.Node(label='A', output=hold_for_t(0.25, 3 * numeral_table[2].v), size_out=D)
    B_in = nengo.Node(label='B', output=hold_for_t(0.25, 3 * numeral_table[3].v), size_out=D)
    # To start cycling
    Push = nengo.Node(label='Push', output=hold_for_t(0.08, 1.))
    # Held Constant
    Fun = nengo.Node(label='Fun', output=_F.v, size_out=D)
    Ref = nengo.Node(label="Ref", output=_ZERO.v, size_out=D)
        
    # Integrators, in general, Fun, and Ref would be as well, 
    # but it involves growing the model too much, KISS etc.
    A, B = constructs.labelled_integrators(['A', 'B'], D, N, positive=True, tau=0.005)

    # Function Application
    NG1 = networks.CircularConvolution(nengo.LIF(N), dimensions=D)
    NG2 = networks.CircularConvolution(nengo.LIF(N), dimensions=D, 
                                       invert_a=True)
    
    # Connect the stuff
    nengo.Connection(A.output, NG1.B)
    nengo.Connection(B.output, NG2.B)
    nengo.Connection(Fun, NG1.A)
    nengo.Connection(Fun, NG2.A)
    nengo.Connection(A_in, A.input)
    nengo.Connection(B_in, B.input)
    
    # Dot-product, should really use the ensemble array, but I can't get it to work.
    Similarity = dot_product.MapReduceDotProduct(leo_neurons=nengo.LIF(n_per_d),
                                                 leo_dimensions=D,
                                                 reducer_neurons=nengo.LIF(2 * n_per_d),
                                                 reducer_radius=0.707)
    # Now we compare 'Ref' to A_prime out of NG1.
    nengo.Connection(NG1.output, Similarity.A)
    nengo.Connection(Ref, Similarity.B)
    
    # The utilities for updating the integrators are based on 1 - similarity
    # The return utility is simply the value of similarity.
    # The idea is once we've become similar to the 'zero' symbol, we're done.
    Q_loop = nengo.Ensemble(neurons=nengo.LIF(n_per_d), dimensions=1, radius=1)
    Q_return = nengo.Ensemble(neurons=nengo.LIF(n_per_d), dimensions=1, radius=1)
    def loop_utility(x): return 1. - x
    nengo.Connection(Similarity.reducer, Q_loop, function=loop_utility)
    nengo.Connection(Similarity.reducer, Q_return)
    nengo.Connection(Push, Q_loop)
    
    # Combine the Utilities into a 2D population for input to BG
    PreBG = nengo.Ensemble(neurons=nengo.LIF(2 * n_per_d), dimensions=2, radius=1)
    nengo.Connection(Q_loop, PreBG, transform=np.array([[1], [0]]))
    nengo.Connection(Q_return, PreBG, transform=np.array([[0], [1]]))
    
    # Now for the Basal Ganglia, which will suppress the input to the integrators
    # This only works because the integrators represent positive values (I think)
    BG = networks.BasalGanglia(dimensions=2, output_weight=-0.5)
    nengo.Connection(BG.output, A.input, transform=np.tile([1, 0], (D, 1)))
    nengo.Connection(BG.output, B.input, transform=np.tile([1, 0], (D, 1)))
    
    # The update connections to integrators:
    nengo.Connection(NG1.output, A.input, synapse=0.005)
    nengo.Connection(NG2.output, B.input, synapse=0.005)
    
    # System Output
    Output = nengo.Node(label="Return", size_in=D, size_out=D)
    nengo.Connection(BG.output, Output, transform=np.tile([0,1], (D, 1)))
    nengo.Connection(B.output, Output)
```

```python
# Model Statistics
total_neurons = 3 * N + 4 * n_per_d + 600
print total_neurons
```

```python
# Simulate and hope for the best?
with model:
    sys_out = nengo.Probe(Output, synapse=0.02)
    sys_similarity = nengo.Probe(Similarity.output, synapse=0.02)
    sys_bg = nengo.Probe(BG.output, synapse=0.02)
    sys_intA = nengo.Probe(A.output, synapse=0.02)
    sys_intB = nengo.Probe(B.output, synapse=0.02)
```

```python
model_sim = nengo.Simulator(model)
model_sim.run(0.5)
```

```python
plt.plot(model_sim.trange(), model_sim.data[sys_out])
```

```python
plt.plot(model_sim.trange(), model_sim.data[sys_similarity])
plt.plot(model_sim.trange(), model_sim.data[sys_bg])
```

```python
f, (a1, a2) = plt.subplots(1, 2, sharey=True)
a1.plot(model_sim.trange(), symbolic_decode(numeral_table, model_sim.data[sys_intA]).T)
a1.set_title("A")
a2.plot(model_sim.trange(), symbolic_decode(numeral_table, model_sim.data[sys_intB]).T);
a2.set_title("B")
```

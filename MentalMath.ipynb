```python
%pylab --no-import-all inline
import numpy as np
import nengo
import symbols
import constructs
import nengo.networks as networks
from nengo.utils.distributions import Uniform
import matplotlib
import matplotlib.pyplot as plt
import dot_product
from integrator import Integrator
#from collections import namedtuple
```

```python
D = 10
n_per_d = 52
N = D * n_per_d
```

```python
# Base Symbols
ZERO = symbols.initial_zero(D)
model = nengo.Network()
numeral_table = dict(symbols.numerals(D, 5, zero=ZERO))
def hold_for_t(t, value):
    '''yields function that outputs value for time <= t'''
    default = np.zeros_like(value)
    def f(tau):
        return value if tau <= t else default
    return f

def symbolic_decode(numeral_table, x):
    similarities = {k: np.dot(v.v, x.T)
             for k, v in numeral_table.iteritems()}
    return np.vstack([d for _, d in
                     sorted([(k, v)
                         for k, v in similarities.iteritems()])])
```

## The Model


```python
with model:
    A_in = nengo.Node(label='A', 
            output=hold_for_t(0.08, numeral_table[1].v), 
            size_out=D)
    B_in = nengo.Node(label='B', 
            output=hold_for_t(0.08, numeral_table[2].v), 
            size_out=D)
    # Held Constant
    Fun = nengo.Node(label='Fun', output=ZERO.v, size_out=D)
        
    # Integrators, in general, Fun, and Ref would be as well, 
    # but it involves growing the model too much, KISS etc.
    A, B = [Integrator(recurrent_tau=0.2, 
                       neurons=nengo.LIF(n_neurons=N),
                       dimensions=D, 
                       radius=1., 
                       max_rates=Uniform(40, 200))
            for i in xrange(2)]
    

    # Function Application
    Aprime = networks.CircularConvolution(nengo.LIF(N), 
                                       dimensions=D)
    Bprime = networks.CircularConvolution(nengo.LIF(N),  
                                       dimensions=D, 
                                       invert_a=True)
    
    # Connect the stuff
    nengo.Connection(A.output, Aprime.B, synapse=0.008)
    nengo.Connection(B.output, Bprime.B, synapse=0.008)
    # simulate coming from integrator
    nengo.Connection(Fun, Aprime.A, synapse=0.008)  
    nengo.Connection(Fun, Bprime.A, synapse=0.008)
    nengo.Connection(A_in, A.input, synapse=None)
    nengo.Connection(B_in, B.input, synapse=None)
    
    Similarity = dot_product.MapReduceDotProduct(
                             leo_neurons=nengo.LIF(n_per_d),
                             leo_dimensions=D,
                             reducer_neurons=nengo.LIF(2 * n_per_d),
                             reducer_radius=1.2)
    # Now we compare 'Ref' to A_prime out of NG1.
    nengo.Connection(Aprime.output, Similarity.A, synapse=0.002)
    
    # The utilities for updating the integrators are based on 1 - similarity
    # The return utility is simply the value of similarity.
    # The idea is once we've become similar to the 'zero' symbol, we're done.
    Q = nengo.Ensemble(neurons=nengo.LIF(n_per_d * 2), 
                       dimensions=2, 
                       radius=0.66)
    def utility(sim):
        '''given similarity  returns [Q_loop, Q_return]'''
        s = sim[0]
        q_loop = np.maximum(0, -10 * s + 1.5)
        q_return = np.maximum(0, 10 * s - 1.5)
        return np.array([q_loop, q_return])
    _intermediary = nengo.Ensemble(neurons=nengo.LIF(n_per_d * 2),
                                   dimensions=2, radius=1)
    nengo.Connection(Similarity.output, _intermediary, 
                     transform=np.ones((2, 1)), synapse=None)
    # Necessary because Pre is a nengo.Node, and therefore can't 
    # be involved in function computing ...
    nengo.Connection(_intermediary, Q, function=utility, synapse=0.002)
    
    
    # Now for the Basal Ganglia, which will suppress the input to the integrators
    # This only works because the integrators represent positive values (I think)
    BG = networks.BasalGanglia(dimensions=2, output_weight=-3)
    nengo.Connection(Q, BG.input, synapse=0.006)
    
    # Generate the 'Gates', these are controlled by the 'iterate' action
    A_gate, B_gate = [dot_product.ThalamusGate(signal, 
                                      inhibit=BG.output,
                                      inhibit_transform=np.array([[1., 0]]),
                                      mapper_neurons=nengo.LIF(n_per_d),
                                      control_neurons=nengo.LIF(n_per_d))
                      for signal in (Aprime.output, Bprime.output)
                      ]
    
    nengo.Connection(A_gate.output, A.input, synapse=0.008)
    nengo.Connection(B_gate.output, B.input, synapse=0.008)
    
    # Ignore the 'return state', too many neurons already ...
    
    # System Output
    Output = nengo.Node(label="Return", size_in=D, size_out=D)
    nengo.Connection(B.output, Output)
```

```python
# Simulate and hope for the best?
with model:
    sys_Q = nengo.Probe(Q, synapse=0.02)
    sys_sim = nengo.Probe(Similarity.output, synapse=0.02)
    sys_bg = nengo.Probe(BG.output, synapse=0.02)
    sys_intA = nengo.Probe(A.output, synapse=0.02)
    sys_intB = nengo.Probe(B.output, synapse=0.02)
    sys_g1 = nengo.Probe(A_gate.output, synapse=0.02)
```

```python
model_sim = nengo.Simulator(model, dt=0.002)  # Holy crap expensive ...
```

```python
model_sim.run(1.)
```

```python
plt.plot(model_sim.trange(), model_sim.data[sys_Q])
plt.legend(['Loop', 'Return'], loc='best')
plt.grid()
plt.title("Utility Scores")
plt.xlabel("t [s]")
plt.ylabel("Q")
```

```python
f, a1 = plt.subplots(1, 1, sharex=True, figsize=(12, 6))
a1.plot(model_sim.trange(), model_sim.data[sys_g1], c='k', alpha=0.7)
a1.set_ylabel("Exhitatory Gated Components")
a1.set_xlabel("t [s]")
a1.legend(['Gated Components'], loc='upper left')
a2 = a1.twinx()
a2.plot(model_sim.trange(), model_sim.data[sys_bg][:,0], label="Inhibitory Control Signal")
a2.set_ylabel("Inhibitory BG Control Signal")
a2.legend(['BG Signal'], loc='upper right')
a1.set_title("Thalamus Gate in Action")
a1.grid()
```

```python
plt.plot(model_sim.trange(), model_sim.data[sys_bg])
plt.legend(["Loop", "Return"], loc='best')
plt.title('Basal Ganglia Output')
plt.xlabel("t [s]")
plt.grid()
```

```python
f, (a1, a2) = plt.subplots(2, 1, sharex=True, figsize=(12, 6))
a1.plot(model_sim.trange(), symbolic_decode(numeral_table, model_sim.data[sys_intA]).T)
a1.set_title("Integrator A, initial value = ONE")
a2.plot(model_sim.trange(), symbolic_decode(numeral_table, model_sim.data[sys_intB]).T);
a2.set_title("Integrator B, initial value = TWO")
plt.legend(["ZERO", 'ONE', 'TWO', 'THREE', 'FOUR'], bbox_to_anchor=(1.1, 1.5))
a2.set_xlabel("t [s]")
a2.set_ylabel("similarity score")
a1.set_ylabel("similarity score")
a1.grid()
a2.grid()
```
